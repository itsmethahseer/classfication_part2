```python
import pandas as pd
df = pd.read_csv('files/weatherAUS.csv')
df.head()
```

Now i can see how much null values are occure in each feature, that's why i can remove the most null value features.

```python
df = df.drop(['Location','Date','Evaporation','Sunshine','Cloud9am','Cloud3pm'],axis=1)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['WindGustDir'] = le.fit_transform(df.WindGustDir)
df['WindDir3pm'] = le.fit_transform(df.WindDir3pm)
df['WindDir9am'] = le.fit_transform(df.WindDir9am)
df['RainToday'] = le.fit_transform(df.RainToday)
df['RainTomorrow'] = le.fit_transform(df.RainTomorrow)

```

```python
df.head()
```

```python
X = df.drop('RainTomorrow',axis='columns')
Y = df.RainTomorrow
```

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'MaxTemp', y = 'MinTemp', hue = 'RainTomorrow' , palette = 'inferno',data = df)
```

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'Humidity9am', y = 'Temp9am', hue = 'RainTomorrow' , palette = 'inferno',data = df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
```

```python
model2.fit(X_train,Y_train)
```

```python
model2.score(X_test,Y_test)
```
