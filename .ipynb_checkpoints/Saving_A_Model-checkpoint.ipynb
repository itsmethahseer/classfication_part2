```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("files/regression.csv")
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
import pickle
```

```python
with open('model.pickle','wb') as f:
    pickle.dump(model,f)
```

Now my model is saved as a file called model.pickle

```python
import pickle
with open('model.pickle','rb') as f:
    m = pickle.load(f)
    
m.predict([[3000]])
```

Now i used my saved model .And predicted without training it again.

##### Both pickle and joblib are good. but if our training dataset contain lots of numpy  array joblib is good.

## Saving model by joblib

```python
import joblib
joblib.dump(model,'model_joblib')
```

```python
m= joblib.load('model_joblib')
m.predict([[4645]])
```
